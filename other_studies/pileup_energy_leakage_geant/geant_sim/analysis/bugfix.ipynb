```python
def get_edep_calo_tail(event:cppyy.gbl.PIMCEvent):
    '''function to get the energy deposited in the gm2 calo xtals and in the tail catcher'''
    deps = {}
    tail_deps = {}
    total = 0
    tail = 0
    # print(event.GetUniqueID())
    for x in event.fCalo:
        name = x.GetCaloID()
        ei = x.GetTotalEnergyDeposit()

        # if(name > 5000):
        #     print(name, ei)
        if(name < 5000):
            ei = x.GetTotalEnergyDeposit()
            total += ei
            deps[name] = ei
        else:
            # if(ei > 500):
            # print(name, ei)
            tail += ei
            tail_deps[name] = ei
    return total, tail, deps, tail_deps
```

```python
loader.SetVerbose(0)
id_interest = 91
for id in range(id_interest+1):
    print("****************************************")
    if(id == id_interest):
        loader.SetVerbose(1)
    else:
        loader.SetVerbose(0)

    event = loader.GetFromFile(id)
    ding = get_edep_calo_tail(event)
    print('->',ding[:2])
```
